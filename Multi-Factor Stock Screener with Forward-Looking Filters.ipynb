```python
tickers = [
    "AAPL", "MSFT", "GOOGL", "AMZN", "BRK-B", "NVDA", "META", "UNH", "XOM", "JNJ",
    "V", "PG", "JPM", "LLY", "TSLA", "MA", "HD", "MRK", "PEP", "AVGO",
    "COST", "ABBV", "KO", "ADBE", "WMT", "CSCO", "CRM", "MCD", "TMO", "BAC",
    "ABT", "DHR", "INTC", "NKE", "AMGN", "CVX", "VZ", "ACN", "MDT", "TXN",
    "NEE", "LIN", "QCOM", "HON", "UNP", "IBM", "PM", "RTX", "MS", "AMAT",
    "LOW", "CAT", "ORCL", "SBUX", "NOW", "INTU", "GE", "BLK", "SPGI", "ISRG",
    "CI", "BA", "LRCX", "TGT", "SCHW", "ADI", "SYK", "ZTS", "MO", "BKNG",
    "GILD", "PLD", "MDLZ", "USB", "CB", "REGN", "DE", "MMC", "PNC",
    "ADP", "SO", "ETN", "GM", "DUK", "AON", "BDX", "CSX", "PGR", "AEP",
    "EW", "VRTX", "NSC", "EOG", "TFC", "SHW", "ICE", "WM", "APD"
]
```

```python
data = {}
for i in tickers:
    print(i)
    ticker = yf.Ticker(i)
    beta = ticker.info[('beta')]
    forward_pe = ticker.info['forwardPE']
    price_to_book = ticker.info['priceToBook']
    cap = ticker.info['marketCap']/1000000000000
    data[i] = {"Beta": beta, "Forward P/E": forward_pe, "Price to Book": price_to_book, "Market Cap": cap}
```

```python
df = pd.DataFrame(data)
df = df.round(2)
df = df.T
df
```

```python
screened = df[(df["Forward P/E"] < 20)&
    (df["Price to Book"] < 5) &
    (df["Beta"] < 1.5) &
    (df["Market Cap"] > 0) # Trillions
]
```
